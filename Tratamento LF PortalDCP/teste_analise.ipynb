```python
import pandas as pd

#### PEGANDO O ARQUIVO ####
print('PEGANDO O ARQUIVO')
teste = pd.read_excel('teste_analise.xlsx')
display(teste)
print('-._.'*30)
#### FIM ####

#### ORDENANDO EM ORDEM CRESCENTE PELA 'COL_A' ####
print('ORDENANDO EM ORDEM CRESCENTE PELA COL_A')
teste.sort_values('COL_A',inplace=True)
display(teste)
print('-._.'*30)
#### FIM ####

#### RESETANDO O INDEX ####
print('RESETANDO O INDEX')
teste.reset_index(drop=True,inplace=True)
display(teste)
print('-._.'*30)
#### FIM ####

#### EXCLUINDO AS LINHAS COM VALORES VAZIOS ####
print('EXCLUINDO AS LINHAS COM VALORES VAZIOS')
teste.dropna(inplace=True)
display(teste)
print('-._.'*30)
#### FIM ####

#### EXCLUINDO AS LINHAS COM VALORES DUPLICADOS EM RELAÇÃO A 'COL_A' ####
print('EXCLUINDO AS LINHAS COM VALORES DUPLICADOS EM RELAÇÃO A COL_A')
teste.drop_duplicates('COL_A',inplace=True)
display(teste)
print('-._.'*30)
#### FIM ####
      
#### RENOMEANDO A COLUNA 'COL_A' ####
print('RENOMEANDO A COLUNA COL_A')
teste.rename(columns={'COL_A':'col_a'},inplace=True)
display(teste)
print('-._.'*30)
#### FIM ####

#### GERANDO UM NOVO ARQUIVO EXCEL ANALISADO ####
print('GERANDO UM NOVO ARQUIVO EXCEL ANALISADO')
teste.to_excel('analise.xlsx', index=False)
print('-._.'*30)
#### FIM ####
```
